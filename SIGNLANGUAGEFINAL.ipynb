```python
import os

DATA_ROOT = "/content/ISL_CSLRT_Corpus/ISL_CSLRT_Corpus"
VIDEO_ROOT = f"{DATA_ROOT}/Videos_Sentence_Level"
GLOSS_FILE = f"{DATA_ROOT}/corpus_csv_files/ISL Corpus sign glosses.csv"
KEYPOINT_DIR = "/content/ISL_keypoints"

os.makedirs(KEYPOINT_DIR, exist_ok=True)

```

```python
import cv2, mediapipe as mp
import numpy as np

mp_holistic = mp.solutions.holistic

def extract_keypoints(video_path):
    cap = cv2.VideoCapture(video_path)
    seq = []

    with mp_holistic.Holistic(model_complexity=2) as holo:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            res = holo.process(rgb)

            pose = np.array([[l.x, l.y, l.visibility] for l in res.pose_landmarks.landmark]) if res.pose_landmarks else np.zeros((33,3))
            lh = np.array([[l.x, l.y, l.visibility] for l in res.left_hand_landmarks.landmark]) if res.left_hand_landmarks else np.zeros((21,3))
            rh = np.array([[l.x, l.y, l.visibility] for l in res.right_hand_landmarks.landmark]) if res.right_hand_landmarks else np.zeros((21,3))

            face = np.zeros((30,3))
            if res.face_landmarks:
                face_all = np.array([[l.x, l.y, l.visibility] for l in res.face_landmarks.landmark])
                face = face_all[:30]

            seq.append(np.concatenate([pose, lh, rh, face], axis=0))

    cap.release()
    return np.array(seq)

```

```python
import glob, tqdm

videos = glob.glob(f"{VIDEO_ROOT}/**/*.mp4", recursive=True)

for v in tqdm.tqdm(videos):
    out = v.replace(VIDEO_ROOT, KEYPOINT_DIR).replace(".mp4",".npy")
    os.makedirs(os.path.dirname(out), exist_ok=True)
    np.save(out, extract_keypoints(v))

```

```python
import pandas as pd
import re

df = pd.read_csv(GLOSS_FILE)
df = df.rename(columns={"Sentence":"sentence","SIGN GLOSSES":"glosses"})
df["glosses"] = df["glosses"].str.split()

def normalize(s):
    s = s.lower().strip()
    s = re.sub(r"[^a-z0-9\s]", "", s)
    return s

df["norm"] = df["sentence"].apply(normalize)

folders = os.listdir(VIDEO_ROOT)
folder_map = {normalize(f):f for f in folders}
df["folder"] = df["norm"].map(folder_map)
df = df[df.folder.notna()].reset_index(drop=True)

```

```python
all_glosses = sorted({g for lst in df["glosses"] for g in lst})
gloss2id = {"<blank>":0}
for i,g in enumerate(all_glosses,1):
    gloss2id[g] = i
id2gloss = {v:k for k,v in gloss2id.items()}
```

```python
def __getitem__(self, i):
    row = self.df.iloc[i]

    files = sorted(glob.glob(os.path.join(self.root, row.folder, "*.npy")))
    if not files:
        return None

    # ---- deterministic selection for validation ----
    kp = np.load(files[0])

    kp = kp[::1]        # temporal downsampling

    T, J, C = kp.shape
    kp = kp.reshape(T, J * C).astype(np.float32)

    gloss = torch.tensor([self.g2i[g] for g in row.glosses])
    return torch.tensor(kp), gloss
```

```python
import torch
import torch.nn as nn

class ISLTransformer(nn.Module):
    def __init__(self, input_dim=315, hidden=512, layers=4, vocab=200):
        super().__init__()
        self.proj = nn.Linear(input_dim, hidden)

        enc = nn.TransformerEncoderLayer(
            d_model=hidden,
            nhead=8,
            batch_first=True
        )
        self.encoder = nn.TransformerEncoder(enc, layers)
        self.fc = nn.Linear(hidden, vocab)

    def forward(self, x, lengths):



        mask = torch.arange(
            x.size(1), device=x.device
        )[None, :] >= torch.tensor(
            lengths, device=x.device
        )[:, None]


        x = self.proj(x)


        x = x[:, ::2]
        mask = mask[:, ::2]


        x = self.encoder(x, src_key_padding_mask=mask)


        return self.fc(x)

```

```python
import torch

def collate(batch):

    batch = [b for b in batch if b is not None]


    if len(batch) == 0:
        return None

    vids, labs = zip(*batch)

    vlen = [v.shape[0] for v in vids]
    llen = [l.shape[0] for l in labs]

    maxv = max(vlen)
    maxl = max(llen)

    feat = vids[0].shape[1]

    pv = torch.zeros(len(vids), maxv, feat)
    pl = torch.zeros(len(labs), maxl, dtype=torch.long)

    for i, (v, l) in enumerate(zip(vids, labs)):
        pv[i, :v.shape[0]] = v
        pl[i, :l.shape[0]] = l

    return pv, pl, vlen, llen
```

```python
import glob
import torch
from torch.utils.data import Dataset
import numpy as np

class ISLDataset(Dataset):
    def __init__(self, df, root, g2i):
        self.df = df
        self.root = root
        self.g2i = g2i

    def __len__(self):
        return len(self.df)

    def __getitem__(self, i):
        row = self.df.iloc[i]

        files = sorted(glob.glob(os.path.join(self.root, row.folder, "*.npy")))
        if not files:
            return None

        # 🔒 deterministic video choice
        kp = np.load(files[0])

        # temporal sampling (same as before)
        kp = kp[::2]

        T, J, C = kp.shape
        kp = kp.reshape(T, J * C).astype(np.float32)

        gloss = torch.tensor([self.g2i[g] for g in row.glosses])

        return torch.tensor(kp), gloss
```

```python
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
import torch
import torch.nn as nn

train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

train_ds = ISLDataset(train_df, KEYPOINT_DIR, gloss2id)
val_ds   = ISLDataset(val_df, KEYPOINT_DIR, gloss2id)

train_loader = DataLoader(
    train_ds,
    batch_size=2,
    shuffle=True,
    collate_fn=collate
)

val_loader = DataLoader(
    val_ds,
    batch_size=2,
    collate_fn=collate
)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = ISLTransformer(vocab=len(gloss2id)).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)
ctc = nn.CTCLoss(blank=0, zero_infinity=True)

EPOCHS = 230

for epoch in range(EPOCHS):
    model.train()
    loss_sum = 0.0
    batch_count = 0

    for batch in train_loader:

        if batch is None:
            continue

        v, l, vl, ll = batch

        v = v.to(device)
        l = l.to(device)


        logits = model(v, vl)
        logp = logits.log_softmax(2).transpose(0, 1)


        vl_ctc = [x // 2 for x in vl]


        loss = ctc(
            logp,
            l,
            torch.tensor(vl_ctc, dtype=torch.long),
            torch.tensor(ll, dtype=torch.long)
        )


        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        optimizer.step()

        loss_sum += loss.item()
        batch_count += 1

    avg_loss = loss_sum / max(1, batch_count)
    print(f"Epoch {epoch+1:03d} | Avg Train Loss: {avg_loss:.4f}")

```

```python
def greedy_decode(lp):
    out = []
    prev = None
    for t in lp.argmax(1):
        t = t.item()
        if t != 0 and t != prev:
            out.append(t)
        prev = t
    return out
```

```python
correct = 0
total = 0

model.eval()
with torch.no_grad():
    for batch in val_loader:
        if batch is None:
            continue

        v, l, vl, ll = batch
        v = v.to(device)

        logits = model(v, vl)
        logp = logits.log_softmax(2)
        logp[:, :, 0] -= 1.0   # your original blank suppression

        for i in range(len(v)):
            pred = greedy_decode(logp[i, :vl[i] // 2])
            truth = l[i, :ll[i]].tolist()

            correct += sum(p == t for p, t in zip(pred, truth))
            total += len(truth)

print("Token Accuracy:", correct / max(1, total))
```

```python
def wer(ref, hyp):
    """
    Token-level Word Error Rate
    """

    ref = list(map(str, ref))
    hyp = list(map(str, hyp))


    import numpy as np

    d = np.zeros((len(ref)+1, len(hyp)+1), dtype=int)

    for i in range(len(ref)+1):
        d[i][0] = i
    for j in range(len(hyp)+1):
        d[0][j] = j

    for i in range(1, len(ref)+1):
        for j in range(1, len(hyp)+1):
            if ref[i-1] == hyp[j-1]:
                cost = 0
            else:
                cost = 1
            d[i][j] = min(
                d[i-1][j] + 1,
                d[i][j-1] + 1,
                d[i-1][j-1] + cost
            )

    return d[len(ref)][len(hyp)] / max(1, len(ref))
```

```python
wers = []

model.eval()
with torch.no_grad():
    for batch in val_loader:
        if batch is None:
            continue

        v, l, vl, ll = batch
        v = v.to(device)

        logits = model(v, vl)
        logp = logits.log_softmax(2)
        logp[:, :, 0] -= 1.0

        for i in range(len(v)):
            pred = greedy_decode(logp[i, :vl[i] // 2])
            truth = l[i, :ll[i]].tolist()
            wers.append(wer(truth, pred))

print("Validation WER:", sum(wers) / len(wers))
```

```python
def set_accuracy(pred, truth):
    return len(set(pred) & set(truth)) / max(1, len(set(truth)))


def length_normalized_accuracy(pred, truth):
    return len(set(pred) & set(truth)) / max(1, len(truth))


def precision(pred, truth):
    return len(set(pred) & set(truth)) / max(1, len(set(pred)))


def recall(pred, truth):
    return len(set(pred) & set(truth)) / max(1, len(set(truth)))


def f1_score(pred, truth):
    p = precision(pred, truth)
    r = recall(pred, truth)
    if p + r == 0:
        return 0.0
    return 2 * p * r / (p + r)


def coverage_accuracy(pred, truth):
    return len(pred) / max(1, len(truth))
```

```python
set_accs = []
len_accs = []
f1s = []
coverages = []

model.eval()
with torch.no_grad():
    for batch in val_loader:
        if batch is None:
            continue

        v, l, vl, ll = batch
        v = v.to(device)

        logits = model(v, vl)
        logp = logits.log_softmax(2)
        logp[:, :, 0] -= 1.0   # DO NOT CHANGE THIS

        for i in range(len(v)):
            pred = greedy_decode(logp[i, :vl[i] // 2])
            truth = l[i, :ll[i]].tolist()

            set_accs.append(set_accuracy(pred, truth))
            len_accs.append(length_normalized_accuracy(pred, truth))
            f1s.append(f1_score(pred, truth))
            coverages.append(coverage_accuracy(pred, truth))
```

```python
print("Set Accuracy:", sum(set_accs) / len(set_accs))
print("Length-Normalized Accuracy:", sum(len_accs) / len(len_accs))
print("F1 Score:", sum(f1s) / len(f1s))
print("Coverage Accuracy:", sum(coverages) / len(coverages))
```
